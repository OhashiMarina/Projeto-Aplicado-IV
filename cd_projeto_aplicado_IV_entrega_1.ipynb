# **Introdução**

A poluição atmosférica é um dos maiores desafios ambientais enfrentados pelas sociedades modernas, especialmente em áreas urbanas e industrializadas. No estado de São Paulo, a concentração populacional e o intenso desenvolvimento econômico têm gerado preocupações crescentes quanto à qualidade do ar e seus impactos na saúde pública e no meio ambiente, conforme destacado pela Organização Mundial da Saúde (OMS) em 2021, que aponta a poluição do ar como responsável por aproximadamente 7 milhões de mortes prematuras a cada ano, sendo a maioria em regiões urbana. Este estudo se insere na área de conhecimento das ciências ambientais e da saúde pública, abordando a questão da poluição do ar como um problema complexo que exige uma análise detalhada e soluções eficazes.



# **Motivações e justificativa**

A problemática da poluição atmosférica em São Paulo é particularmente relevante, pois afeta diretamente a saúde de milhões de pessoas e contribui para a degradação ambiental. Em 2020, a Environmental Protection Agency (EPA) dos Estados Unidos relatou que a exposição prolongada a poluentes atmosféricos está associada a um aumento de doenças respiratórias e cardiovasculares. Além disso, o problema está intrinsecamente ligado aos Objetivos de Desenvolvimento Sustentável (ODS) da Organização das Nações Unidas (ONU), especificamente ao ODS 13, que trata da Ação contra a Mudança Global do Clima. A poluição do ar, ao intensificar os efeitos das mudanças climáticas, compromete a sustentabilidade ambiental e o bem-estar das futuras gerações conforme divulgado pela Companhia Ambiental do Estado de São Paulo (Cetesb) em 2021.

Este trabalho pretende explorar a situação atual da poluição atmosférica paulista. A análise buscará identificar tendências e padrões sazonais além de prever concentrações de poluentes para os próximos anos. Ao abordar essas questões, espera-se contribuir não apenas para a compreensão dos desafios ambientais enfrentados no estado de São Paulo, mas também para o desenvolvimento de políticas públicas que estejam alinhadas com os ODS e que promovam uma transição para um futuro mais sustentável.

Assim, o estudo sobre a poluição atmosférica em São Paulo não apenas evidencia um problema de grande relevância local e global, mas também destaca a necessidade de uma ação coordenada para enfrentar os desafios ambientais do século XXI, em conformidade com os compromissos internacionais assumidos pelo Brasil.

Hora: hora em que foi feita a medição;

Estação: local em que foi feita a medição;

Código: código associado à estação em que foi realizada a medição;

Poluente: poluente cuja concentração foi medida;

Valor: valor, na unidade especificada, da concentração do poluente;

Unidade: unidade de concentração utilizada; e

Tipo: como foi realizada a medição (de forma automática ou manual).
