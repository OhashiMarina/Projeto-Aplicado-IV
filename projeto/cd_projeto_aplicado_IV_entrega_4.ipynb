# **Referencial Teórico**

A análise da qualidade do ar é uma questão de grande relevância mundial devido aos impactos diretos na saúde pública e no meio ambiente. A poluição atmosférica, caracterizada pela presença de substâncias nocivas como material particulado (PM), dióxido de nitrogênio (NO₂), ozônio (O₃) e monóxido de carbono (CO), tem sido associada a diversos problemas de saúde, incluindo doenças respiratórias e cardiovasculares (SALDIVA et al., 2012). Estudos indicam que longos períodos de exposição a poluentes aumentam o risco de mortalidade prematura, especialmente em populações vulneráveis, como crianças e idosos (WHO, 2018). Além disso, os impactos econômicos são expressivos, com custos associados ao tratamento de doenças relacionadas à poluição e à perda de produtividade.

No Brasil, a cidade de São Paulo enfrenta desafios significativos no controle da qualidade do ar, devido ao seu tráfego intenso, crescimento populacional e atividades industriais. Por esse motivo, a análise da qualidade do ar se torna crucial para avaliar a eficácia das políticas públicas implementadas, além de permitir a antecipação de medidas preventivas em episódios de alta poluição. A compreensão de tendências e padrões sazonais na concentração de poluentes, como o CO, auxilia na tomada de decisões para mitigar seus efeitos adversos e contribui para o desenvolvimento de estratégias de controle mais eficazes (BRAGA et al., 2001).

A aplicação de técnicas de modelagem de séries temporais e métodos preditivos para analisar a qualidade do ar tem se tornado uma abordagem amplamente utilizada na literatura.

Um exemplo é o estudo de Liu et al. (2021), que utilizou o modelo SARIMA (Seasonal Autoregressive Integrated Moving Average) para prever a concentração de material particulado em grandes cidades da China. O modelo SARIMA se mostrou eficaz na captura de sazonalidades e tendências temporais, confirmando sua aplicabilidade em cenários de poluição atmosférica com variações sazonais (LIU et al., 2021).

Outro trabalho relevante, de Chen e Huang (2020), investigou o uso de redes neurais recorrentes (RNN) e Long Short-Term Memory (LSTM) para prever concentrações de NO₂ em grandes centros urbanos. Os resultados demonstraram que as redes neurais, especialmente LSTM, foram capazes de capturar padrões não lineares nas séries temporais de poluentes, oferecendo previsões mais precisas em comparação aos métodos tradicionais (CHEN; HUANG, 2020).

Esses modelos são particularmente relevantes para o presente trabalho, dado que os objetivos incluem tanto a identificação de padrões sazonais quanto a modelagem preditiva das concentrações de poluentes na cidade de São Paulo. Ao adotar técnicas de modelagem de séries temporais, como SARIMA e LSTM, pretende-se prever as futuras concentrações de poluentes com base em dados históricos, gerando informações relevantes para a implementação de políticas de mitigação de poluição e a adaptação das cidades às mudanças ambientais.

A baixa umidade dificulta a dispersão dos poluentes, resultando em níveis mais altos de CO. #Teste Dickey-Fuller

```python
from statsmodels.tsa.stattools import adfuller
def teste_estacionaria(timeseries):

    # Determinando estatísticas móveis
    rolmean = timeseries.rolling(12).mean() # média móvel
    rolstd = timeseries.rolling(12).std() # desvio médio

    # Plot de estatísticas móveis
    orig = plt.plot(timeseries, color='blue', label="Original")
    mean = plt.plot(rolmean, color="red", label="Média móvel")
    std = plt.plot(rolstd, color="black", label="Désvio médio")
    plt.legend(loc="best")
    plt.title("Média móvel e Desvio Padrão")
    plt.show(block=False)


    # Perform Dickey-Fuller test:
    print("Resultados do teste de Dickey-Fuller")
    dftest = adfuller(timeseries, autolag="AIC")

    # Organizando resultados
    dfoutput = pd.Series(dftest[0:4], index=["Estatística do teste", "p-value", "O critério de informação maximizado", "Número de observações usadas"])
    for key, value in dftest[4].items():
        dfoutput[f"Valor crítico ({key})"]= value

    print(dfoutput)
```

```python
def estacionaria(concentracao_ano):
    dftest = adfuller(concentracao_ano, autolag="AIC")
    dfoutput = pd.Series(dftest[0:4], index=["Estatítica do teste", "p-value", "O critério de informação maximizado", "Número de observações usadas"])
    for key, value in dftest[4].items():
         dfoutput["Valor crítico (%s)"%key] = value

    if(dfoutput["Estatítica do teste"] < dfoutput["Valor crítico (5%)"] and dfoutput["p-value"] < 0.05):
        print("É estacionária")
    else:
        print("Não é estacionária")
```

# **Modelos**

Neste estudo, foram utilizados três modelos distintos para prever a concentração de monóxido de carbono (CO) ao longo do ano de 2022, cada qual com seus hiperparâmetros ajustados.

O primeiro modelo, SARIMA (1,1,1)(0,1,1,12), foi escolhido por sua capacidade de lidar com séries temporais não estacionárias e com sazonalidade mensal, características observadas nos dados históricos analisados. Os hiperparâmetros foram ajustados com order (p,d,q) igual a (1,1,1), para capturar componentes autorregressivas (AR), integração (I) e médias móveis (MA), e seasonal order (P,D,Q,s) igual a (0,1,1,12), que reflete a sazonalidade mensal. Após o treinamento, o modelo gerou previsões para os 12 meses do ano de 2022.

O segundo modelo SARIMA utilizou hiperparâmetros ajustados para order (p,d,q) igual a (2,1,2) e seasonal order (P,D,Q,s) igual a (0,1,1,12). Adicionalmente, foi empregada a técnica de validação cruzada com divisão temporal (TimeSeriesSplit) em cinco divisões, permitindo uma análise mais robusta do desempenho do modelo ao longo do tempo. Essa abordagem possibilitou avaliar o modelo em múltiplos períodos futuros, treinando-o em dados históricos de diferentes intervalos temporais.

Por fim, foi implementado um modelo de rede neural LSTM (Long Short-Term Memory), que oferece uma abordagem não linear e mais flexível para séries temporais. Os hiperparâmetros incluíram 12 passos históricos como entrada, 50 neurônios na camada LSTM, função de ativação ReLU, função de perda MSE e otimizador Adam. O treinamento foi realizado em 50 épocas com batches de tamanho 16, utilizando 80% dos dados para treino e 20% para teste. Os dados foram previamente normalizados para o intervalo [0,1], facilitando a convergência do modelo.

# Modelo 3 - Aplicação do modelo LSTM

```python
# Importando bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Carregando os dados
data = concentracao_ano.copy()
data.index = pd.to_datetime(data.index)
values = data['Valor'].values.reshape(-1, 1)

# Normalizando os dados para o intervalo [0,1]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_values = scaler.fit_transform(values)

# Criando as sequências para treinamento
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i + n_steps])
        y.append(data[i + n_steps])
    return np.array(X), np.array(y)

n_steps = 12  # Número de meses para considerar no histórico
X, y = create_sequences(scaled_values, n_steps)

# Dividindo os dados em treino e teste (80%/20%)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Construindo o modelo LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Treinando o modelo
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=0)

# Avaliando o modelo
loss = model.evaluate(X_test, y_test)
print(f"Loss no conjunto de teste: {loss}")

# Fazendo previsões
y_pred = model.predict(X_test)

# Transformando de volta para os valores originais
y_pred_inverse = scaler.inverse_transform(y_pred)
y_test_inverse = scaler.inverse_transform(y_test)

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.plot(range(len(y_test_inverse)), y_test_inverse, label='Valores Reais')
plt.plot(range(len(y_pred_inverse)), y_pred_inverse, label='Previsões', linestyle='--')
#plt.title('Previsão de Concentração de CO2 usando LSTM')
plt.xlabel('Tempo')
plt.ylabel('Concentração de CO2')
plt.legend()
plt.show()
```

```python
# Aavaliando o modelo LSTM
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse}')

# Calculando outras metricas
from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")


# Plotando a validacao e treinamento
plt.plot(history.history['loss'], label='Valores de perda no treinamento')
plt.plot(history.history['val_loss'], label='Valores de perda na validação')
#plt.title('Valores de perda no treinamento e validação')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# **Discussão e Conclusão**

Este estudo apresentou a implementação de diferentes modelos preditivos para estimar a concentração de monóxido de carbono (CO) ao longo de 2022, utilizando abordagens estatísticas e baseadas em aprendizado de máquina. Os modelos SARIMA demonstraram robustez ao lidar com séries temporais sazonais, capturando tendências e sazonalidades de maneira satisfatória. Por fim, uma análise mais aprofundada do desempenho em diferentes horizontes temporais poderia trazer insights sobre o comportamento dos modelos em previsões de curto e longo prazo, aumentando sua aplicabilidade prática.
