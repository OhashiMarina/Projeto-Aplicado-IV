# **Modelo base**

Foi realizado o teste de Dickey-Fuller Aumentado (ADF) para verificar a estacionariedade da série temporal de concentração de CO. A estatística do teste obtida foi -0,787, com um p-value de 0,822, indicando que não há evidências suficientes para rejeitar a hipótese nula de não estacionariedade. Além disso, a estatística do teste é maior que todos os valores críticos (1%, 5% e 10%), reforçando que a série apresenta comportamento não estacionário. Portanto, conclui-se que a média e a variância da concentração de CO não permanecem constantes ao longo do tempo, o que pode exigir transformações ou diferenciações para modelagem adequada.

Utilizamos o modelo SARIMA (1,1,1)(0,1,1,12) para realizar a previsão da concentração de CO para os 12 meses do ano de 2022. A escolha desse modelo levou em conta tanto a componente não estacionária identificada quanto a sazonalidade mensal presente nos dados. Após ajustar o modelo com os dados históricos, foram geradas previsões mensais para todo o ano de 2022, permitindo uma análise antecipada dos níveis de poluição e auxiliando no planejamento de políticas públicas de mitigação.

A avaliação do modelo SARIMA foi realizada utilizando a métrica Root Mean Squared Error (RMSE), que mede a diferença entre os valores previstos e os valores reais. O RMSE obtido foi 1.215.894, indicando um erro significativo na precisão das previsões. Esse valor sugere que, apesar do ajuste do modelo, ele não conseguiu capturar completamente as variações da concentração de CO ao longo do período analisado. Possíveis causas para essa discrepância incluem a natureza não estacionária da série e a presença de padrões sazonais complexos. Como aprimoramento futuro, pode-se considerar um ajuste mais refinado dos hiperparâmetros.

```python
from statsmodels.tsa.stattools import adfuller
def teste_estacionaria(timeseries):

    # Determinando estatísticas móveis
    rolmean = timeseries.rolling(12).mean() # média móvel
    rolstd = timeseries.rolling(12).std() # desvio médio

    # Plot de estatísticas móveis
    orig = plt.plot(timeseries, color='blue', label="Original")
    mean = plt.plot(rolmean, color="red", label="Média móvel")
    std = plt.plot(rolstd, color="black", label="Désvio médio")
    plt.legend(loc="best")
    plt.title("Média móvel e Desvio Padrão")
    plt.show(block=False)


    # Perform Dickey-Fuller test:
    print("Resultados do teste de Dickey-Fuller")
    dftest = adfuller(timeseries, autolag="AIC")

    # Organizando resultados
    dfoutput = pd.Series(dftest[0:4], index=["Estatística do teste", "p-value", "O critério de informação maximizado", "Número de observações usadas"])
    for key, value in dftest[4].items():
        dfoutput[f"Valor crítico ({key})"]= value

    print(dfoutput)
```

```python
def estacionaria(concentracao_ano):
    dftest = adfuller(concentracao_ano, autolag="AIC")
    dfoutput = pd.Series(dftest[0:4], index=["Estatítica do teste", "p-value", "O critério de informação maximizado", "Número de observações usadas"])
    for key, value in dftest[4].items():
         dfoutput["Valor crítico (%s)"%key] = value

    if(dfoutput["Estatítica do teste"] < dfoutput["Valor crítico (5%)"] and dfoutput["p-value"] < 0.05):
        print("É estacionária")
    else:
        print("Não é estacionária")
```

```python
# Ajuste do modelo SARIMA
modelo_sarima = SARIMAX(concentracao_ano['Valor'], order=(1,1,1), seasonal_order=(0,1,1,12))
```

```python
# Treinando o modelo
resultado = modelo_sarima.fit()
```

```python
# Prevendo os poluentes para os próximos 12 meses (ano inteiro de 2022)
previsoes_2022 = resultado.forecast(steps=12)
print(previsoes_2022)
```
